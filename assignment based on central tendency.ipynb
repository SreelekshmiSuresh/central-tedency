```python
#3. Who scores the most on average for math, reading and writing based on
#● Gender
# ● Test preparation course
```

```python
#● Gender
```

```python
df1 = data1[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').mean()
df1
```

```python
# ● Test preparation course
```

```python
df2= data1[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').mean()
df2
```

```python
# 4. What do you think about the scoring variation for math, reading and writing
# based on
# ● Gender
# ● Test preparation course
```

```python
# ● Gender
```

```python
df1 = data1[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').std()
df1
```

```python
# ● Test preparation course
```

```python
df2= data1[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').std()
df2
```
